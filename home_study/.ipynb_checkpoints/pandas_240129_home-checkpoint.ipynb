```python
ebola=pd.read_csv('./do_it_pandas-main/data/country_timeseries.csv')
ebola.columns
```

```python
ebola.info()
```

```python
ebola.iloc[:5,[0,1,2,10]]
```

```python
ebola_long=ebola.melt(id_vars=['Date','Day'])
ebola_long
```

```python
variable_split=ebola_long.variable.str.split('_')
variable_split[:5]
```

```python
status_values=variable_split.str.get(0)
country_values=variable_split.str.get(1)
```

```python
country_values
```

```python
ebola_long['status']=status_values
ebola_long['country']=country_values
ebola_long
```

```python
ebola_long=ebola.melt(id_vars=['Date','Day'])
ebola_long
```

```python
variable_split=ebola_long.variable.str.split('_', expand=True)
variable_split
```

```python
ebola_long[['status','country']]=variable_split
ebola_long
```

```python
weather=pd.read_csv('./do_it_pandas-main/data/weather.csv')
weather.head()
```

```python
weather_melt=weather.melt(id_vars=['id','year','month','element'], var_name='day',value_name='temp')
weather_melt.head()
```

```python
weather_tidy=weather_melt.pivot_table(index=['id','year','month','day'],
                                      columns='element', values='temp')
weather_tidy.info()
```

```python
weather_tidy_flat=weather_tidy.reset_index()
weather_tidy_flat[:10]
```

```python
weather_tiny=(
    weather_melt.pivot_table(index=['id','year','month','day'], 
                             columns='element',values='temp').reset_index())
weather_tiny[:10]
```
