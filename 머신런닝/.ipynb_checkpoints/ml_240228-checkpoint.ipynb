```python
# 통계적 접근
# 성별에 따른 비용: 정규성T검정, 등분산성 
from scipy import stats

s0= df[df['sex']=='female']['charges']
s1= df[df['sex']=='male']['charges']

print(stats.shapiro(s0))
print(stats.shapiro(s1))
```

```python
# 주 집단 모두 정규성을 가지지 않으므로 비모수 검정 시행
stats.mannwhitneyu(s0,s1)
# 연관이 너무 없으므로 뺄지 말지 결정해도 됨
```

```python
# 흡연여부: 비용에 따른
s0= df[df['smoker']=='yes']['charges']
s1= df[df['smoker']=='no']['charges']
print(stats.shapiro(s0))
print(stats.shapiro(s1))
```

```python
stats.mannwhitneyu(s0,s1)   # 대립가설 채택: 엄청 상관이 있다
```

```python
# 지역: F검정

s0= df[df['region']=='northeast']['charges']
s1= df[df['region']=='southeast']['charges']
s2= df[df['region']=='southwest']['charges']
s3= df[df['region']=='northwest']['charges']

print(stats.shapiro(s0))
print(stats.shapiro(s1))
print(stats.shapiro(s2))
print(stats.shapiro(s3))
```

```python
# 정규분포 아니므로 비모수 검정을 한다
stats.kruskal(s0,s1,s2,s3)
```

```python
# f검정 -분산 : 거의 대립가성 채캑할 뻔함
stats.f_oneway(s0,s1,s2,s3)
```

```python
# child 비용 관계
s0= df[df['child']==0]['charges']
s1= df[df['child']==1]['charges']
s2= df[df['child']==2]['charges']
print(stats.shapiro(s0))
print(stats.shapiro(s1))
print(stats.shapiro(s2))
```

```python
stats.kruskal(s0,s1,s2)   # 대립가설 채택: 자녀수에따라 비용 달라진다.
```
