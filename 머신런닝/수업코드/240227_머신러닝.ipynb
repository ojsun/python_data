```python
# 데이터 분석 이 끝나면
# 머신러닝 분석을 진행하면 된다.
# 데이터 전처리 / 모델 선택 / 학습 / 예측 / 평가
```

```python
# 데이터 전처리 - 필요없는 컬럼 제거(Time)
credit_res = credit_res.drop('Time', axis = 1)
```

```python
# 데이터 전처리 - 스케일링(Amount)
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
credit_res['Amount_scaled'] = scaler.fit_transform(credit_res[['Amount']])
credit_res = credit_res.drop('Amount', axis = 1)
```

```python
# 데이터 전처리 - 학습/테스트 데이터 분할
from sklearn.model_selection import train_test_split
y = credit_res['Class']
X = credit_res.drop('Class', axis = 1) # v1 ~ v28, amount_scaled

X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size = 0.2, 
                                                    random_state = 0)

print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
# 모델 선택
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from xgboost import XGBClassifier

dt_clf = DecisionTreeClassifier(random_state=0)
rf_clf = RandomForestClassifier(random_state=0)
gbm_clf = GradientBoostingClassifier(random_state=0)
knn_clf = KNeighborsClassifier()
lr_clf = LogisticRegression(random_state=0)
svm_clf = SVC(random_state=0)
xgb_clf = XGBClassifier(random_state=0)
```

```python
scores = cross_val_score(dt_clf, X_train, y_train, scoring='accuracy', cv = 3)
```

```python
# pipeline
# 데이터 전처리 + 학습/예측/평가
# diabetes
diabetes = pd.read_csv('data/ml/diabetes.csv')
diabetes.head()
```

```python
# X, y
X = diabetes.iloc[:, :-1]
y = diabetes.iloc[:, -1]

# train / test
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, 
                                                    random_state=0)

# -----------------------------------
# 이후 작업을 pipeline으로 묶기
# StandardScaler
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 학습
from sklearn.linear_model import LogisticRegression

lr_clf = LogisticRegression(random_state=0)
lr_clf.fit(X_train_scaled, y_train)

# 예측
pred = lr_clf.predict(X_test_scaled)
# ------------------------------------------------------

# 평가
from sklearn.metrics import accuracy_score

print(accuracy_score(y_test, pred))
```

```python
# 파이프라인 구축 - (스케일러, 모델)
from sklearn.pipeline import Pipeline

scaler = StandardScaler()
lr_clf = LogisticRegression(random_state=0)

pipe = Pipeline(  [ ('SC', scaler), ('LR', lr_clf) ]   )
```

```python
# 파이프라인 학습 - (스케일링, 모델학습)
pipe.fit(X_train, y_train)
```

```python
# 파이프라인 예측 - (스케일링, 모델예측)
pred = pipe.predict(X_test)
```

```python
print(accuracy_score(y_test, pred))
```

```python
# Pipeline + GridSearchCV

# GridSearchCV(lr_clf, param_grid={'C':[0.01, 0.1]})
params = {'LR__C':[0.01, 0.1]}
grid = GridSearchCV(pipe, param_grid=params, scoring='accuracy', cv=3)

grid.fit(X_train, y_train)
```
