### Step 5. 데이터에서 한국 KOR 데이터만 추출해주세요.

```python
df_kr=df[df['Country']=='KOR']
df_kr
```

### Step 6. 한국 올림픽 메달리스트 데이터에서 년도에 따른 medal 종류별 갯수를 데이터프레임화하세요.
* 메달이 없으면 0으로 나오게 하세요.

```python
df_kr.groupby(['Year','Medal']).size().to_frame().reset_index()
```

```python
df_kr.pivot_table(index='Year', columns='Medal', aggfunc='size', fill_value=0).reset_index()
```

### Step 7. 전체 데이터에서 sport종류에 따른 성별수를 구하세요.

```python
df.pivot_table(index='Sport', columns='Gender', aggfunc='size').reset_index()
```

### Step 8. 전체 데이터에서 Discipline종류에 따른 따른 Medal수를 구하세요.

```python
df.pivot_table(index='Discipline', columns='Medal', aggfunc='size').reset_index()
```

### Step 9. 한국 올림픽 데이터에서 금메달을 가장 많이 획득한 Discipline와 Event의 종류를 구하세요.

```python
df_kr_g=df_kr[df_kr['Medal']=='Gold']
df_kr_g
```

```python
df_kr_g.pivot_table(index=['Discipline','Event'], columns='Medal',aggfunc='size').sort_values('Gold', ascending=False).reset_index()
```
