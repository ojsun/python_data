```python
titanic=pd.read_csv('../data/titanic.csv')
titanic.head()
```

```python
titanic1=titanic[['Survived','Age']].dropna()
t0=titanic1[titanic1['Survived']==0]
t1=titanic1[titanic1['Survived']==1]

plt.subplot(1,2,1)
sns.kdeplot(x='Age', data=t0, label='Died')
sns.kdeplot(x='Age', data=t1, label='Survived')
plt.legend()

plt.subplot(1,2,2)
sns.histplot(x='Age', data=titanic1, hue='Survived', kde=True)
```

 대립가설: 생존여부(범주형)에 따른 나이(수치형)의 차이가 있다.
1. 범주(집단)을 나누어서 분석 시작해야 함: 나이 데이터로
2. 결측치 처리- 제거나 채우기 (결측치가 있으면 작동 안함)

```python
s0 = titanic[titanic['Survived'] == 0]['Age'].dropna()
s1 = titanic[titanic['Survived'] == 1]['Age'].dropna()
plt.subplot(1,2,1)
sns.histplot(s0, kde=True)
plt.subplot(1,2,2)
sns.histplot(s1, kde=True)
plt.tight_layout()
```

```python
sns.histplot(data=titanic, x='Age', hue='Survived', kde=True )
```

1. 두집단의 정규성 검정: shapiro/anderson/kstest
1_1. 정규성을 만족했을때  : 독립성 확인  --> 등분산성 검정 
1_2. 정규성을 불만족했을때  --> 맨-휘트니 유 검정  mannwhitneyu

2. 독립성 검정
2_1. 독립성 만족(대부분 만족): 할 필요없음 --> 등분산성
2_2. 독립성 불만족: 대응표본 t검정

3. 등분산성 검정
3_1 등분산성을 만족했을때: 독립표본 t검정( 등분산 True)
3_2.등분산성을 불만족했을때: 독립표본 t검정( 등분산 False)

```python
# 1_1. 정규성 검정 (전체 AGE 검정)
# shapiro() 데이터 5000미만/ anderson() 5000이상/ kstest()

stats.shapiro(titanic['Age'].dropna())
```

```python
stats.anderson(titanic['Age'].dropna())
```

```python
stats.kstest(titanic['Age'].dropna(), 'norm')
```

```python
# 1_2.집단별 정규성 검정
print(stats.shapiro(s0))
print(stats.shapiro(s1))
```

```python
# 두 집단 모두 정규성 갖지 않음 --> 맨 휘트니 유 검정 (비모수검정)
stats.mannwhitneyu(s0,s1)

# p-value가 0.05보다 크므로 두 집단 간의 평균차이가 없다. 
# 집단간 평균차이가 없다는 귀무가설 기각 안함.
```

```python
# 2. 정규성만족할 경우 
# --> 독립성 확인 (두 집단이 요소가 겹치는 것이 있는지): 할 필요가 없음 --> 대부분 독립적
```

```python
# 3. 독립성 만족할 경우 
# --> 등분산성 검정: bartlett/ levene / fligner   귀무: 두 집단이 등분산을 가진다.
```

```python
print(stats.bartlett(s0, s1))
print(stats.levene(s0, s1))
print(stats.fligner(s0, s1))

# p-value가 0.05보다 크므로 귀무가설이 기각되지 않는다.(등분산을 가진다)
```

```python
# 3_1 등분산성을 만족했을때: 
# 독립표본 t검정( 등분산 True)/ 등분산 안 가질때 (False) 귀무: 두 집단의 평균차이가 없다.
stats.ttest_ind(s0,s1, equal_var=True)

# p-value가 0.05가 작으므로 귀무가설을 기각 
# 생존여부에 따라 --> 평균차이가 있다.
```

```python
# 3_2 등분산성을 만족했을때:
stats.ttest_ind(s0,s1, equal_var=False)
```

```python
p1 = titanic[titanic['Pclass'] == 1]['Age'].dropna()
p2 = titanic[titanic['Pclass'] == 2]['Age'].dropna()
p3 = titanic[titanic['Pclass'] == 3]['Age'].dropna()
sns.barplot(x = 'Pclass', y = 'Age', data = titanic)
```

```python
# 일원배치 분산분석 : F검정, ANOVA 검정
# 정규성0, 독립성 0, 등분산성 0
stats.f_oneway(p1,p2,p3)

# 귀무: 집단간의 분산의 차이가 없다
# p-value가 0.05보다 적으므로 귀무가설을 기각 - 차이가 있다.
# 즉, 객석 등급에 따른 나이차이가 있다.
```

```python
# 정규성을 가지 않으때 집단이 3개 이상인 경우
stats.kruskal(p1,p2,p3)
# 분산 차이 있다
```
