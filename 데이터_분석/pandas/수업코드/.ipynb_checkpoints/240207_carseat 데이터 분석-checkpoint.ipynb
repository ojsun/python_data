```python
# 범주vs수치(판매량), T or F
# ShelveLoc, US

# barplot(평균비교)
sns.barplot(data = carseat, x = 'ShelveLoc', y = 'Sales')
```

```python
car_g = carseat.groupby('ShelveLoc')['Sales'].get_group('Good')
car_m = carseat.groupby('ShelveLoc')['Sales'].get_group('Medium')
car_b = carseat.groupby('ShelveLoc')['Sales'].get_group('Bad')
```

```python
# 정규성
print(stats.shapiro(car_g))
print(stats.shapiro(car_m))
print(stats.shapiro(car_b))
```

```python
# 정규성o, 독립성o, 등분산성 검정
print(stats.bartlett(car_g, car_m, car_b))
print(stats.levene(car_g, car_m, car_b))
print(stats.fligner(car_g, car_m, car_b))
```

```python
# f_oneway ANOVA 검정
stats.f_oneway(car_g, car_m, car_b)
```

```python
sns.barplot(data = carseat, x = 'US', y = 'Sales')
```

```python
car_y = carseat[carseat['US'] == 'Yes']['Sales']
car_n = carseat[carseat['US'] == 'No']['Sales']
```

```python
print(stats.shapiro(car_y))
print(stats.shapiro(car_n))
```

```python
sns.histplot(car_y, kde = True)
```

```python
# 정규성o, 독립성o, 등분산성
print(stats.bartlett(car_y, car_n))
print(stats.levene(car_y, car_n))
print(stats.fligner(car_y, car_n))
```

```python
stats.ttest_ind(car_y, car_n, equal_var=True)
```

```python
stats.ttest_ind(car_y, car_n, equal_var=False)
```

```python
# 선반 위치 여부에 따라서 판매량의 차이가 있다.
# 국내, 해외 여부 따라서 판매량의 차이가 있다.
```

```python
car_ad0 = carseat[carseat['Advertising'] == 0]['Sales']
car_ad1 = carseat[carseat['Advertising'] != 0]['Sales']
```

```python
plt.bar( [ 'ad0', 'ad1'  ]  , [car_ad0.mean(), car_ad1.mean()])
```

```python
print(stats.shapiro(car_ad0))
print(stats.shapiro(car_ad1))
sns.histplot(car_ad0, kde = True)
```

```python
# 정규성o, 독립성o, 등분산성
print(stats.bartlett(car_ad0, car_ad1))
```

```python
stats.ttest_ind(car_ad0, car_ad1)
```
