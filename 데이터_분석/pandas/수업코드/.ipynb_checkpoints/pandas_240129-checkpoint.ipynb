```python
avg_lifeexp_by_year=df.groupby('year')['lifeExp'].mean()
# DD에서의 사용법 select avg(lifeExp) from df group by year
avg_lifeexp_by_year
```

```python
def my_mean(values):
    n = len(values)
    sum = 0
    for value in values:
        sum += value
    return sum / n
```

```python
agg_my_mean=df.groupby('year')['lifeExp'].apply(my_mean)
agg_my_mean
```

```python
def my_mean_diff(values, diff_value):
    n = len(values)
    sum = 0
    for value in values:
        sum += value
    mean = sum / n
    return mean - diff_value
```

```python
global_mean = df['lifeExp'].mean()
```

```python
agg_mean_diff = df.groupby('year')['lifeExp'].agg(my_mean_diff, diff_value=global_mean)
agg_mean_diff
```

```python
#agg없이 게산하기
df.groupby('year')['lifeExp'].mean() - global_mean   # 검토할것
```
