```python
stats.pearsonr(air['Temp'], air['Ozone']) # 귀무가설 : 연관성이 없다
# p-value가 0.05보다 작으면 귀무가설 기각, 대립가설 채택(두 변수 상에 연관성이 있다.)
# 연관성이 있으니까 검정통계량 값이 0.68로 상관계수를 가진다.

# 온도에 따른 오존 수치가 0.68 정도 되는 상관성을 가진다.
```

```python
# 귀무가설 : 두 변수간의 연관성이 없다.

# pvalue가 0.05보다 작다 > 귀무가설 기각 > 두 변수간의 연관성이 있다.
# 대립가설 채택 > 검정통계량이 맞다.

# pvalue가 0.05보다 크다 > 귀무가설 채택 > 두 변수간의 연관성이 없다.
# 검정통계량이 거짓값이다. 즉 쓸수 없다.
```

```python
titanic = pd.read_csv('data/pandas/titanic.csv')
pd.crosstab(titanic['Survived'], titanic['Sex'])
```

```python
sns.countplot(data = titanic, x = 'Survived', hue = 'Sex')
```

```python
ct = pd.crosstab(titanic['Survived'], titanic['Sex'], normalize='index')
ct.plot.bar(stacked=True) 
```

```python
table = pd.crosstab(titanic['Survived'], titanic['Sex'])
table
```

```python
result = stats.chi2_contingency(table)
print('카이제곱통계량 :', result[0])
print('P-value :', result[1])
print('기대빈도 :\n', result[3])
```

```python
table = pd.crosstab(titanic['Survived'], titanic['Pclass'])
table
```

```python
result = stats.chi2_contingency(table)
print('카이제곱통계량 :', result[0])
print('P-value :', result[1])
print('기대빈도 :\n', result[3])
```

```python
# 머신러닝
# 종속변수 수치형 : 회귀분석, 범주형 : 분류분석
# 독립변수 여러개
```

```python
# 타이타닉 = 모집단
# 성별에 따른 나이값을 표본 (남성, 여성)
age_mean = titanic['Age'].mean() # 모집단의 평균
female_age = titanic[titanic['Sex'] == 'female']['Age'].dropna() # 표본1
male_age = titanic[titanic['Sex'] == 'male']['Age'].dropna()     # 표본2
```

```python
# 표본의 평균이 믿을만 한가?
stats.ttest_1samp(female_age, age_mean) # 단일표본 t검정, 수치형, + 비교대상(모평균)
# 귀무가설 : female_age이 집단의 평균과 age_mean 값이 차이가 없다.(같다.)
# 대립가설 : female_age의 평균과 age_mean 값이 차이가 있다.(다르다.)
```

```python
stats.ttest_1samp(male_age, age_mean)
```

```python
print(age_mean)
print(female_age.mean())
print(male_age.mean())
```

```python
# 생존 여부(범주)에 따른 나이(수치)의 차이가 있다.
s0 = titanic[titanic['Survived'] == 0]['Age'].dropna()
s1 = titanic[titanic['Survived'] == 1]['Age'].dropna()
plt.subplot(1,2,1)
sns.histplot(s0, kde=True)
plt.subplot(1,2,2)
sns.histplot(s1, kde=True)
```

```python
print('사망자 나이')
print(stats.shapiro(s0))
print('생존자 나이')
print(stats.shapiro(s1))
# 두 집단 모두 정규성을 가지지 않으므로 비모수 검정을 진행해야한다.
```

```python
stats.mannwhitneyu(s0, s1) # 귀무가설 : 두 집단의 평균 차이가 없다.
# p-value가 0.05크므로 귀무가설을 채택
# s0, s1 집단은 같은 집단이라고 할수 있다. 
# > 사망자 집단과 생존자 집단의 나이 평균 차이가 없다.
```

```python
# 범주일때 수치형 > 범주형 데이터끼리 집단으로 나눠줘야 한다.
# 집단 2개 > ttest
titanic[titanic['Survived'] == 0]
titanic[titanic['Survived'] == 1]
```

```python
# 집단 3개 > ANOVA
titanic[titanic['Pclass'] == 1]
titanic[titanic['Pclass'] == 2]
titanic[titanic['Pclass'] == 3]
```
