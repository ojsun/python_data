### 1. (정규분포를 가지도록) price와 area 변수의 이상치 제거 후2,3번 문제풀이

```python
# 사분위수를 이용하여 이상치 제거

IQR=h_price.price.quantile(0.75) - h_price.price.quantile(0.25)
high_b=h_price.price.quantile(0.75) + IQR*1.5
low_b=h_price.price.quantile(0.25) - IQR*1.5

h_price1=h_price[(h_price['price']>low_b) & (h_price['price'] <high_b)]

h_price1.head()

# 전체 545 row 중 15개 제거됨
```

```python
h_price.head()
```

```python
sns.boxplot(data=h_price1, x='price')
```

```python
# 그래프로 확인
fig=plt.figure(figsize=(6,4))
plt.subplot(2,1,1)
sns.histplot(data=h_price1, x='price')
plt.subplot(2,1,2)
sns.boxplot(data=h_price1, x='price')
plt.tight_layout()
```

```python
# 정규성 검사
print(stats.shapiro(h_price1['price']))   # 비정규적 분포
```

```python
# 독립변수 area 이상치 제거 -- zscore 이용

z=np.abs(h_price['price'] - h_price['price'].mean())/h_price['price'].std()
h_price3=h_price[z < 2]

# 545 row 중 26 제거됨
```

```python
# 정규성 검사
print(stats.shapiro(h_price3['area']))   # 비정규적 분포
```

##### <결론> 이상치를 제거하고 정규성 검정을 했으나 두 변수 모두 비정규분포를 가지므로 차후 분석은 위와 동일할 것이라 예측됨

### 2. 종속변수 price(수치형) 와 독립변수 bathrooms(범주형) 분석

```python
# price 다음으로 상관관계가 있어보이는 bathroom
# 데이터 통계량 분석

h_price.bathrooms.unique()
```

```python
h_price.bathrooms.value_counts()
h_price2=h_price[h_price['bathrooms'] != 4]

# bathroom 4는 한개의 데이터만 가지고 있는으므로 이상치로 판단, 제거함
```

```python
# 3개의 그룹으로 나누기
bath1=h_price2[h_price2.bathrooms==1]['price']
bath2=h_price2[h_price2.bathrooms==2]['price']
bath3=h_price2[h_price2.bathrooms==3]['price']
```

```python
# 정규성 검사
print('bath1 검증결과')
print(stats.shapiro(bath1))
print('bath2 검증결과')
print(stats.shapiro(bath2))
print('bath3 검증결과')
print(stats.shapiro(bath3))

# p-value 결과로 볼때 bath1,2는 비정규, bath3은 정규로 보이나 표본수가 10개이므로 확정할 수 없음
```

```python
# 그래프로 확인
fig=plt.figure(figsize=(8,7))
plt.subplot(2,1,1)
sns.histplot(data=h_price2, x='price', hue='bathrooms', kde=True)
plt.title('histplot of bathrooms')
plt.subplot(2,1,2)
sns.barplot(y='price',x='bathrooms', data=h_price2)
# 그래프로 볼때, bath3도 정규분포를 따르지않는 것으로 보임
```

```python
p=h_price2.bathrooms.value_counts()
plt.pie(p.values, labels=p.index, autopct='%.2f%%')
```

```python
# 등분산 검정  : 귀무가설-등분산을 갖는다.
print(stats.bartlett(bath1,bath2,bath3))
print(stats.levene(bath1,bath2,bath3))
print(stats.fligner(bath1,bath2,bath3))

# 모든 p-value 결과값이 0.05보다 작으므로 분산이 갖지 않다.
```

```python
# 비정규 분포  --> 크푸스칼 웰리스 테스트
stats.kruskal(bath1,bath2,bath3)

# 귀무가설: 차이가 없다  --> 귀무가설 기각 
# p-value가 0.05보다 작고 통게값이 크므로  각 집단은 분산의 차이가 크다고 볼 수 있다.
# bathroom 갯수에 따라 price에 영향이 있다.
```

```python
# 3개 이상 집단이 등분산 아닐때 웰치스 ANOVA 검정 : 정규성을 가져야 하는데, 그냥 해봄
import pingouin as pg

h_price4=h_price[['bathrooms','price']]
pg.welch_anova(data=h_price4,dv='price',between='bathrooms')

# p-value가 0.05보다 작으므로 귀무가설을 기각/ 대립가설 채택
# 집단 간의 분산의 차이가 있음/ bathroom 갯수에 따라 price에 영향이 있다.
```

### 3. 종속변수 price(수치형) 와 독립변수 mainroad(범주형) 분석

```python
# 통계분석- 독립변수 mainroad

print(h_price2.mainroad.unique())
print(h_price2.mainroad.value_counts())
# yes와 no로 이루어져있으며, 누락값이 없고 yes가 다수를 이루고 있음
```

```python
# 그래프로 확인
fig=plt.figure(figsize=(8,7))
plt.subplot(2,1,1)
sns.histplot(data=h_price2, x='price', hue='mainroad', kde=True)
plt.title('histplot of mainroad')
plt.subplot(2,1,2)
sns.barplot(y='price',x='mainroad', data=h_price2)
```

```python
p=h_price2.mainroad.value_counts()
plt.pie(p.values, labels=p.index, autopct='%.2f%%')
```

```python
# 2개의 그룹으로 나누기
main_y=h_price2[h_price2.mainroad=='yes']['price']
main_n=h_price2[h_price2.mainroad=='no']['price']
```

```python
# 정규성 검사
print('mainroad_yes 정규성 검증결과')
print(stats.shapiro(main_y))
print('mainroad_no 정규성 검증결과_1')
print(stats.shapiro(main_n))
print('mainroad_no 정규성 검증결과_2')
print(stats.anderson(main_n))

# 'yes'의 경우 pvalue가 작아 귀무가설 기각/ 'no'의 경우 0.05보다는 크지만 
# 표본수가 상대적으로 작고 'yes'의 값도 정규적이지 않아 정규분포로 보기 어려움
```

```python
# 등분산 검정  : 귀무가설-등분산을 갖는다.
print('mainroad의 등분산 검정')
print(stats.bartlett(main_y, main_n))
print(stats.levene(main_y,main_n))
print(stats.fligner(main_y,main_n))

# 모든 p-value 결과값이 0.05보다 작으므로 분산이 갖지 않다.
```

```python
# 비정규적이므로 비모수검정 Mann-Whitney U test 검정

stats.mannwhitneyu(main_y,main_n)

# 귀무가설: 두 집단의 중앙값 차이가 없다 즉, 같다 
# p-value가 0.05보다 작으므로 귀무가설을 기각/ 대립가설 채택
```

```python
# 참고: 정규성 갖는 종속(수치), 독립(범주2)의 독립표본 T검정

stats.ttest_ind(main_y,main_n, equal_var=False)

# 기무가설( 두 집단의 평균의 유의마한 차이가 없다) 기각 --> 차이가 있다.
```

####<결론>
집단간의 중앙값 및 평균 등의 차이가 있으므로 mainroad의 유무가 price에 영향이 없다고 할 수 없다.
