### Step 5. 데이터에서 한국 KOR 데이터만 추출해주세요.

```python
kr = df[df['Country'] == 'KOR']
```

### Step 6. 한국 올림픽 메달리스트 데이터에서 년도에 따른 medal (종류별)갯수를 데이터프레임화하세요.
* 메달이 없으면 0으로 나오게 하세요.

```python
kr.pivot_table(index = 'Year', columns = 'Medal',aggfunc='size',)
#               fill_value=0)
```

```python
kr.groupby(['Year', 'Medal']).size().to_frame().reset_index().pivot(index='Year', columns='Medal', values = 0).fillna(0)
```

### Step 7. 전체 데이터에서 sport종류에 따른 남녀별로 수를 구하세요.

```python
df.pivot_table(index='Sport', columns = 'Gender', values = 'Medal', aggfunc='count')
```

### Step 8. 전체 데이터에서 Discipline종류에 따른 따른 Medal수를 구하세요.

```python
df.pivot_table(index='Discipline', columns = 'Medal', aggfunc='size')
```

### Step 9. 한국 올림픽 데이터에서 금메달을 가장 많이 획득한 Discipline와 Event의 종류를 구하세요.

```python
kr.head()
```

```python
kr.pivot_table(index = ['Discipline', 'Event'], columns='Medal', aggfunc='size').sort_values('Gold', ascending=False)
```
