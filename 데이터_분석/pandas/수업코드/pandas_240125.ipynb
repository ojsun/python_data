### merge 로 합치기

```python
person=pd.read_csv('../data/survey_person.csv')
site=pd.read_csv('../data/survey_site.csv')
visited=pd.read_csv('../data/survey_visited.csv')
survey=pd.read_csv('../data/survey_survey.csv')
```

```python
display(person)
```

```python
display(site)
```

```python
display(survey)
```

```python
sv_merge=site.merge(visited, left_on='name', right_on='site')
sv_merge
```

```python
'''select * from site join visited on site.name =visited.site'''
sv_merge=site.merge(visited, left_on='name', right_on='site')
```

```python
sv_merge
```

```python
sp_merge=person.merge(survey, left_on='ident', right_on='person' )
```

```python
sp_merge
```

```python
spsv=sp_merge.merge(sv_merge, left_on='taken', right_on='ident', how='left')
```

```python
spsv  # 다 합친 테이블
```

```python
site2=site.replace('DR-3', 'DR-1')
```

```python
site2.merge(visited, left_on='name', right_on='site', how='outer')  # 중복값 문제 : 동일값 있음
```
