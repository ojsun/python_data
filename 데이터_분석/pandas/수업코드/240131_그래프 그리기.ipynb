```python
f_tips = tips['total_bill']
IQR = f_tips.quantile(q = 0.75) - f_tips.quantile(q = 0.25)

high_border = f_tips.quantile(q = 0.75) + (IQR * 1.5)
low_border = f_tips.quantile(q = 0.25) - (IQR * 1.5)

sns.boxplot(data = tips, y = 'total_bill')
plt.axhline(high_border)
plt.axhline(low_border)
plt.axhline(tips['total_bill'].min())
plt.axhline(tips['total_bill'].max())
```
