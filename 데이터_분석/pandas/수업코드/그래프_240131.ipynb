### mathplotlib 일변량(단변량) - plot(라인), 히스토그램

```python
plt.figure(figsize=(6,8))
plt.plot('total_bill','y*',data=tips.sort_values('total_bill',ignore_index=True), label='bill($)')
plt.title("Plot-라인그래프")
plt.xlabel('total_bill')
plt.ylabel('frequency')
plt.grid(axis='y')
plt.legend()
plt.show()
```

```python
tips['total_bill'].unique().sum()
```

```python
# histgram 히스토그램
plt.figure(figsize=(6,8))
plt.subplot(2,1,1)
plt.hist(tips['total_bill'],bins=70,color='red')

plt.title('Histogram of total bill_70')
plt.xlabel('total_bill')
plt.ylabel('frequency')
plt.subplot(2,1,2)
hist=plt.hist(tips['total_bill'],bins=10)
plt.title('Histogram of total bill_10')
plt.xlabel('total_bill')
plt.ylabel('frequency')
plt.tight_layout()   # 간격조절
plt.show()
```

```python
series=[]
for group in tips.groupby('sex')['total_bill']:
    series.append(group[1])
series
```

```python
box1=plt.boxplot(series, labels=['Mail','Female'], vert=False)
plt.title('Boxplot df total bill by Gender')
plt.grid()
plt.xlabel('Gender')
plt.ylabel('total_bill')
```

# 사분위 알아내기 (선생님)
f_tips = tips[tips['sex'] == 'Male']['total_bill'].describe()
IQR = f_tips['75%'] - f_tips['25%']
high = f_tips['75%'] + IQR * 1.5

box = plt.boxplot(tips[tips['sex'] == 'Male']['total_bill'])
print(box['whiskers'][1].get_ydata())
print(high)

# median - IQR * 1.5 = low border    ==> 나: Q3 + IQR * 1.5 = high border 

```python
# vert=False 일 경우에는 get_xdata 사용해야
print(box1['whiskers'][0].get_xdata()) # 아래쪽 max, min
print(box1['whiskers'][1].get_xdata()) # 위쪽 min, max
print(series[0].describe())
```
